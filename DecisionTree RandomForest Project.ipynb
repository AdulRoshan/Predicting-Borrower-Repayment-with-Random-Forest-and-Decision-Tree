```python
loan['purpose'].value_counts()
```

```python
cat_purpose = pd.get_dummies(loan['purpose'],drop_first=True)
cat_purpose.head()
```

```python
d_final=pd.concat([loan,cat_purpose],axis=1).drop('purpose',axis=1)
d_final
```

```python
cat_feats=['purpose']
```

```python
data_final=pd.get_dummies(loan,columns=cat_feats,drop_first=True) 
data_final
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=d_final.drop('not.fully.paid',axis=1)
y=d_final['not.fully.paid']
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

```python
pred_tree=dtree.predict(X_test)
```

```python
from sklearn import metrics
```

```python
print('CONFUSION MATRIX')
print(metrics.confusion_matrix(y_test,pred_tree))
print('\n')
print('CLASSIFICATION REPORT')
print(metrics.classification_report(y_test,pred_tree))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc= RandomForestClassifier(n_estimators=350)
```

```python
rfc.fit(X_train,y_train)
```

```python
pred_rfc=rfc.predict(X_test)
```

```python
print('CONFUSION MATRIX')
print(metrics.confusion_matrix(y_test,pred_rfc))
print('\n')
print('CLASSIFICATION REPORT')
print(metrics.classification_report(y_test,pred_rfc))
```
